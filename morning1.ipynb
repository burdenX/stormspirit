## 0. python 环境与版本

在选择最新版本的Anaconda时，请选择python3.6版本。现在众多工具包已经宣布停止对python2.7版本的维护和升级，转向python3.6。我们也需要与时俱进（python 2.7可以不用安装了）。我们本节课的重点就是利用anaconda装新的python环境

```python
import sklearn.datasets as datasets
import pandas as pd
iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
y=iris.target
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

### numpy
numpy定义了python进行矩阵数值计算的基础
> np.add(A,B)

> np.subtract(A,B)

> np.dot(C,D)

> np.matmul(C,D)           ## C@D

> np.multiply(C,D)         ## C*D

```python
import numpy as np
A = np.array([1,2,3])
B = np.array([4,3,0])
print('A+B = ',np.add(A,B))
print('A-B = ',np.subtract(A,B))
```

```python
C = np.array([[1,2,3],
            [1,1,1]])
D = np.array([[4,3,9,9],
            [1,1,2,2,],
            [4,2,1,1]])
print('np.dot(C,D) = ',np.dot(C,D))
print('np.matmul(C,D) = ',np.matmul(C,D))
print('        C@D    = ',np.matmul(C,D))

print('np.multiply(D,D)=',np.multiply(D,D))
print('          D*D   = ', D*D)
```

### 重点看看 np.dot 和 np.matmul的区别

* 都是矩阵乘法
* np.matmul中，禁止矩阵与标量的乘法
* np.matmul中，在矢量乘矢量的内积运算中，matmul与dot没有差别
* np.matmul中，多维的矩阵，将前n-2维视为后2维的元素后，进行乘法操作

#### 矢量内积无区别，结果是标量

```python
t1 = np.array([1,2,3])
t2 = np.array([1,2,3])
print(np.dot(t1,t2))
print(np.matmul(t1,t2))
print(t1*t2) # 元素1 1 乘积
```

#### 1维与2维矩阵的乘法

```python
a = np.array([1,2,3])
b = np.array([[1,1,1],
             [2,2,2],
             [3,3,3]])
```

```python
np.dot(a,b)
```

```python
np.matmul(a,b)
```

```python
#注意
a*b # 扩展维度后，元素乘积
```

#### 多维矩阵乘积的比较

```python
m_4_4 = np.array([
                       [1,2,3,4],
                       [3,2,1,4],
                       [5,4,6,7],
                       [11,12,13,14]
                      ])


m_3_4_2 = np.array([
                             [[2,3],
                              [11,9],
                              [32,21],
                              [28,17]],
    
                             [[2,3],
                              [1,9],
                              [3,21],
                              [28,7]],
    
                             [[2,3],
                              [1,9],
                              [3,21],
                              [28,7]]
                            ])
```

```python
print(m_4_4.shape)
print(m_3_4_2.shape)
```

```python
print('4x4*3x4x2 dot:\n {}\n'.format(np.dot(m_4_4,m_3_4_2)))
print('4x4*3x4x2 matmul:\n {}\n'.format(np.matmul(m_4_4,m_3_4_2)))
```

```python
print('4 4 矩阵 与')
print('3 4 2 矩阵的乘积结果')
print('dot 结果的维度    =',np.dot(m_4_4,m_3_4_2).shape)
print('matmul 结果的维度 =',np.matmul(m_4_4,m_3_4_2).shape)
```

**检验以下m_4_4与m_3_4_2最后两维的矩阵乘积结果**

```python
np.dot(m_4_4,m_3_4_2[0,:,:])
```

```python
np.matmul(m_4_4,m_3_4_2[:,:,0].T)
```
